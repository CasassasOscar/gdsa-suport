## Feature extraction with Bag of Words

The first thing we need to do is to generate a single descriptor for every image in the training and validation sets. As we explained in earlier sessions, we will use the Bag of Words aggregation technique, which consists in training a codebook of visual words and building an image descriptor encoding the number of times each word in the codebook appears in the image.

### Step 1: Local feature extraction with the training set

To build the codebook we only use the images in the training set, thus we need to compute local descriptors for all training images and stack them together in a single numpy array. The function *stack_features* does exactly this procedure.

```python
import get_features as GF
```

```python
# Make sure that we are using training images only !
params['split'] = 'train'

t = time.time()
X, pca, scaler = GF.stack_features(params)

print "Done. Time elapsed:", time.time() - t
print np.shape(X)
```

Now, *X* contains all the local features of the training set. Notice that even though we significantly reduced the size of the images (width = 300 and keeping the aspect ratio), this took a few minutes to run, and we ended up with ~200.000 descriptors. Using bigger images is going to be slower and will generate more descriptors (but results will probably be a lot better !). Notice how the function *stack_features* also returns the parameters *pca* and *scaler*, which would contain the transformations that we applied to X (e.g. dimensionality reduction, whitening, centering, ...), in case we chose to do that. In this example I am not applying any feature normalization or transformation, so these two variables are *None*.

### Step 2: Training the codebook

Now that we have the training features ready, we can train our codebook. In my code I ended up using the [MiniBatchKMeans](http://scikit-learn.org/stable/modules/generated/sklearn.cluster.MiniBatchKMeans.html) instead of the regular [KMeans](http://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html) class from [scikit-learn](http://scikit-learn.org/stable/). MiniBatchKMeans is much faster for large amounts of training descriptors (> 10.000), and we have more than 200.000!

The function *train_codebook* saves the model with the trained clusters to disk using *pickle*.

```python
t = time.time()
GF.train_codebook(params,X)

print "Done. Time elapsed:", time.time() - t
```

That was fast ! Training with the same descriptors in *X* using KMeans would have been a lot slower, trust me. In any case, a higher number of centroids  and/or local features will result in higher computational time in both cases.

## Generating Rankings

At this point we have extracted and stored BoW descriptors for all images in the training and validation sets. Now, we just need to compute distances of each image descriptor in the validation set to all image descriptors in the training set. This will give us the dissimilarity between each query image of the validation set to each image in the training set. Once we have these distances, we can use them to sort the images in the training set in descending order. This sorted list is the final ranking for the query image.

To do this, I have updated last week's *rank* function, in order to take into account the distances to create the ranking instead of just sorting the training images randomly.
