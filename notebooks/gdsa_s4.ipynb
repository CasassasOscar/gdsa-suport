## Random Classification

We start with our classification results. What we show here are the results of our random classifier, which assigned a label (out of the 13 possible ones) to each validation image. 

```python
import eval_classification as EC

# Get metrics
f1, precision,recall, accuracy, cm, labels = EC.eval_classification(params)
```

Perhaps the most useful metric in this case is the Confusion Matrix. It gives us an idea of how well our random classifier performs for each one of the classes. Ideally, we would like to have a confusion matrix with high values in the diagonal, and 0 elsewhere. This would mean that all images are being correctly classified. But because we chose the labels randomly, we are not even close to that ideal scenario.

```python
EC.plot_confusion_matrix(cm,labels,normalize=False)
```

Notice that the values in the "desconegut" row are higher than the rest. This is only due to the fact that that class has more samples than the rest. This may be a bit confusing, so we can simply normalize the rows of the matrix by the number of samples in each class to better understand what is going on.

```python
EC.plot_confusion_matrix(cm,labels,normalize=True)
```

Better? At this point you might be thinking why aren't all values in the matrix slightly the same. If we picked the class labels randomly, all classes have the same probablity, and the confusion matrix should be flat (e.g. all positions should have nearly the same value). The reason why this does not happen is that our validation set only contains 10 images for each building, and our random classifier can choose between 13 classes. There are more possible classes than images per class. Actually, if you look at the "desconegut" row in the normalized confusion matrix you can see that it looks more even than the rest, and again it's because this class has 6 times more images, so it's more likely that all classes will be picked with the same probability.

We can also take a look at the rest of the measures:

```python
print "F1 per class:"
print zip(labels,f1)
print "Precision per class:"
print zip(labels,precision)
print "Recall per class:"
print zip(labels,recall)
```

```python
print "Average F1:", np.mean(f1)
print "Average Precision:", np.mean(precision)
print "Average Recall:", np.mean(recall)
print "Accuracy:", accuracy
```

Yep, it's pretty bad. But we kind of knew it beforehand.
