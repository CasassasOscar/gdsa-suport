## Step 1: Train the classifier

Since we already have the BoW descriptors computed and stored, we can jump to training our classifier right on.

```python
import src.classify as CL
```

```python
t = time.time()

CL.train_classifier(params)

print "Done. Time elapsed:", time.time() - t
```

The function *train_classifier* reads the BoW features for the training data and trains an SVM classifier with it. Instead of choosing the default parameters, I have created a function that tries several parameter configurations (the ones in *params['svm_tune']*) and picks the combination that gives a higher F1 measure. Finally, the model is saved to disk using *pickle*.

## Step 2: Classification

Now that we have our trained classifier, we can use it to predict the class of our validation images.

```python
t = time.time()
CL.classify(params)
print "Done. Time elapsed:", time.time() - t
```

The function *classify* calls the *predict()* method of the SVM classifier and returns the class with a highest score. Finally, it saves a text file with all the class predictions of the validation images.

## Step 3: Evaluation

Now it's time to evaluate our classification results. We will use the same functions as we did in S3.

```python
import src.eval_classification as EC
```

```python
f1, precision, recall, accuracy, cm, labels = EC.eval_classification(params)

print zip(labels,f1)
print "==============================="
print "F1:", np.mean(f1)
print "Precision:", np.mean(precision)
print "Recall:", np.mean(recall)
print "Accuracy:", np.mean(accuracy)
```

This is a huge improvement from our random results in S3 ! Let's look at the confusion matrix...

```python
EC.plot_confusion_matrix(cm,labels,normalize = True)
```

It looks like it's working well, though it seems that most of the mistakes are due to confusion with the *desconegut* class. In other terms, it means that everytime the classifier is not sure, it predicts the *desconegut* class. This is kind of reasonable if you think about it, since this class has many more samples than the rest, and the probability of success is higher.
