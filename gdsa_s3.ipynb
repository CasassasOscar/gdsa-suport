This notebook shows an example of how to use a keypoint detector to identify and describe local areas of an image.

```python
import numpy as np
import os, sys, time
import cv2
from utils.drawmatches import drawMatches
```

```python
# Read images
im1 = cv2.imread('./images/eet0.jpg')
im2 = cv2.imread('./images/eet1.jpg')

print np.shape(im1), np.shape(im2)
```

```python
# Initialize SIFT detector
sift = cv2.SIFT()

keypoints1, descriptors1 = sift.detectAndCompute(im1,None)
keypoints2, descriptors2 = sift.detectAndCompute(im2,None)

print np.shape(keypoints1), np.shape(keypoints2)
```

We can display the keypoints found by the SIFT detector as blobs in the image. The descriptor of each keypoint is computed in the area within the blob.

```python
# Display keypoints on images
imshow(cv2.drawKeypoints(cv2.cvtColor(im1,cv2.COLOR_BGR2RGB),keypoints1))
show()
imshow(cv2.drawKeypoints(cv2.cvtColor(im2,cv2.COLOR_BGR2RGB),keypoints2))
show()
```

As we computed keypoints and descriptors for two images of the same instance (Escola d'Enginyeria de Terrassa), we can try to match keypoints from one image to the other. To do this, we can use the Brute Force Matcher from the OpenCV library. See http://docs.opencv.org/master/dc/dc3/tutorial_py_matcher.html#gsc.tab=0 for more details on Feature Matching.

```python
# Going further: Matching keypoints from the two images

bf = cv2.BFMatcher()
matches = bf.match(descriptors1,descriptors2)
matches = sorted(matches,key=lambda val: val.distance)

print len(matches)
```

Notice that we get some good matches, but a high number of outliers as well. Spatial verification techniques can be used to detect potential outliers.

```python
# Show the matches
out = drawMatches(cv2.cvtColor(im1,cv2.COLOR_BGR2GRAY), keypoints1, cv2.cvtColor(im2,cv2.COLOR_BGR2GRAY), keypoints2, matches[:20])

imshow(out)
show()
```
